```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
N, J  = 2000, 1
sigma = np.sqrt(2J*J/N)

mat = np.random.normal(loc=0, scale=sigma, size=(N, N))
for i in range(N):
    for j in range(i):
        mat[i, j] = mat[j, i]
```

```python
eigen_values = np.linalg.eig(mat).eigenvalues
```

```python
plt.hist(eigen_values, bins=30, density=True)

sigma_p = np.max(eigen_values)/2

lambda_X = np.linspace(-sigma_p*2, sigma_p*2)
p_lambda = np.sqrt(4*sigma_p**2 - lambda_X**2)/(2*np.pi*sigma_p**2)

plt.plot(lambda_X, p_lambda)
plt.show()
```

# Task 2

```python
sorted_eigen_values = np.sort(eigen_values)
eigen_value_step = sorted_eigen_values[1:] - sorted_eigen_values[:-1]
eigen_value_step /= np.mean(eigen_value_step)

x_lim = np.percentile(eigen_value_step, 99)
plt.hist(eigen_value_step, bins=np.linspace(0, x_lim, 30), density=True)

step_X = np.linspace(0, x_lim, 100)
p_step = (np.pi*(step_X)/2)*np.exp(-np.pi*(step_X**2)/4)

plt.plot(step_X, p_step, "r")
plt.show()
```

```python
N, J  = 2000, 1
sigma = np.sqrt(2J*J/(N/2))

mat1, mat2 = np.random.normal(loc=0, scale=sigma, size=(N//2, N//2)), np.random.normal(loc=0, scale=sigma, size=(N//2, N//2))
for i in range(N//2):
    for j in range(i):
        mat1[i, j] = mat1[j, i]
        mat2[i, j] = mat2[j, i]
```

```python
eigen_values = np.array([np.linalg.eigvals(mat1), np.linalg.eigvals(mat2)])
```

```python
plt.hist(np.reshape(eigen_values, (-1, )), bins=30, density=True)

sigma_p = np.max(eigen_values)/2

lambda_X = np.linspace(-sigma_p*2, sigma_p*2)
p_lambda = np.sqrt(4*sigma_p**2 - lambda_X**2)/(2*np.pi*sigma_p**2)

plt.plot(lambda_X, p_lambda)
plt.show()
```

```python
sorted_eigen_values = np.sort(eigen_values, axis=1)
eigen_value_step    = sorted_eigen_values[:, 1:] - sorted_eigen_values[:, :-1]
eigen_value_step   /= np.expand_dims(np.mean(eigen_value_step, axis=1), 1)

x_lim = np.percentile(np.reshape(eigen_value_step, (-1, )), 99)
plt.hist(np.reshape(eigen_value_step, (-1, )), bins=np.linspace(0, x_lim, 30), density=True)

step_X = np.linspace(0, x_lim, 100)
p_step = (np.pi*(step_X)/2)*np.exp(-np.pi*(step_X**2)/4)

plt.plot(step_X, p_step, "r")
plt.show()
```
